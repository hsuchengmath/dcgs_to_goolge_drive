```python
import pulp
ProbLP1 = pulp.LpProblem("ProbLP1", sense=pulp.LpMaximize)
```

```python
x1 = pulp.LpVariable('x1', lowBound=0, upBound=15, cat='Integer')
x2 = pulp.LpVariable('x2', lowBound=0, upBound=7.5, cat='Integer')
ProbLP1 += (10*x1 + 9*x2)
ProbLP1 += (6*x1 + 5*x2 <= 62) 
ProbLP1 += (10*x1 + 20*x2 <= 150)
pulp.LpSolverDefault.msg = None
ProbLP1.solve()
```

```python
print(ProbLP1.name)  # 輸出求解狀態
print("Status:", pulp.LpStatus[ProbLP1.status])  # 輸出求解狀態
for v in ProbLP1.variables():
    print(v.name, "=", v.varValue)  # 輸出每個變數的最優值
print("F1(x) =", pulp.value(ProbLP1.objective))  # 輸出最優解的目標函式值
```
